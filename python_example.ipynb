```python
# Load some modules
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr

%matplotlib inline # this is only for demonstration, you don't need this line in your code.
```

```python
# Start with defining a height array
z = np.linspace(1,300,20) # an array from 1 to 300 in 20 steps
z # to print the content of z below
```

```python
# Proceed with defining some useful constants
k = 0.4
z0 = 0.002
ust = 0.25
```

```python
# Compute the wind speed at each altitude
u = ust/k*np.log(z/z0)
u # to print the content of u below
```

```python
# Visualization:
fig,ax = plt.subplots() # this creates the figure and a set of axes
ax.plot(u,z) # this plots u versus z on the set of axes (called 'ax')
plt.show()
```

```python
# Let's do that again but tweak the axes a bit
fig,ax = plt.subplots() # this creates the figure and a set of axes
ax.plot(u,z) # this plots u versus z on the set of axes (called 'ax')

# Set axis limits
ax.set_xlim(0,10)
ax.set_ylim(0,300)

# Set axis labels
ax.set_xlabel('wind speed (m/s)')
ax.set_ylabel('altitude (m)')

plt.show()
```

```python
# Now, let's make a function that calculates the logprofile
def logprof(ust,z0,z):
    ''' This function calculates u given ust, z0 and a range of z values'''
    kappa = 0.4 # vond karman constant
    u = ust/kappa*np.log(z/z0)
    return u

# And for convenience also make a function that plots it
def plotprof(u,z,ax):
    ''' This function plots u versus z on a given set of axes ax '''
    ax.plot(u,z)
    return

# Check if it works:
fig,ax = plt.subplots() # create a new figure instance

z = np.linspace(1,300,20)

# Let's plot 2 profiles together
u = logprof(0.25,0.0002,z)
plotprof(u,z,ax)

u = logprof(0.35,0.0002,z)
plotprof(u,z,ax)

plt.show()
```
